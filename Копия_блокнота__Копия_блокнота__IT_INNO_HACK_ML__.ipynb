```python
import pandas as pd
import re
from fuzzywuzzy import fuzz
from sklearn.metrics import precision_score, recall_score, f1_score

# Загрузка датасетов
dataset1 = pd.read_csv('main1.csv', nrows=30)
dataset2 = pd.read_csv('main2.csv', nrows=30)
dataset3 = pd.read_csv('main3.csv', nrows=30)

# Очищаем и стандартизируем данные для первого датасета
dataset1.columns = dataset1.columns.str.strip().str.lower()  # Приведение названий столбцов к нижнему регистру
dataset1['full_name'] = dataset1['full_name'].str.strip().str.lower()
dataset1['email'] = dataset1['email'].str.strip().str.lower()
dataset1['address'] = dataset1['address'].str.strip().str.lower()

# Стандартизация телефонов
dataset1['phone'] = dataset1['phone'].str.replace(r'\D', '', regex=True)  # Удаляем все нечисловые символы
dataset1['phone'] = dataset1['phone'].apply(lambda x: f"+7 (0{x[1:4]}) {x[4:7]}-{x[7:9]}-{x[9:11]}" if len(x) >= 11 else x)

# Очищаем и стандартизируем данные для второго датасета
dataset2.columns = dataset2.columns.str.strip().str.lower()  # Приведение названий столбцов к нижнему регистру
dataset2['first_name'] = dataset2['first_name'].str.strip().str.lower()
dataset2['middle_name'] = dataset2['middle_name'].str.strip().str.lower()
dataset2['last_name'] = dataset2['last_name'].str.strip().str.lower()
#dataset2['email'] = dataset2['email'].str.strip().str.lower()
dataset2['address'] = dataset2['address'].str.strip().str.lower()

# Стандартизация телефонов
dataset2['phone'] = dataset2['phone'].str.replace(r'\D', '', regex=True)  # Удаляем все нечисловые символы
dataset2['phone'] = dataset2['phone'].apply(lambda x: f"+7 (0{x[1:4]}) {x[4:7]}-{x[7:9]}-{x[9:11]}" if len(x) >= 11 else x)

# Очищаем и стандартизируем данные для третьего датасета
dataset3.columns = dataset3.columns.str.strip().str.lower()  # Приведение названий столбцов к нижнему регистру
dataset3['name'] = dataset3['name'].str.strip().str.lower()
dataset3['email'] = dataset3['email'].str.strip().str.lower()
dataset3['birthdate'] = pd.to_datetime(dataset3['birthdate'], errors='coerce').dt.date  # Приведение формата даты
dataset3['sex'] = dataset3['sex'].str.strip().str.lower()  # Приведение пола к нижнему регистру

# Вывод первых строк очищенных датасетов
print("Dataset 1:")
print(dataset1)
print("\nDataset 2:")
print(dataset2)
print("\nDataset 3:")
print(dataset3)
```

## Предообработка данных

```python
# Шаг 3: Метод блокировки (Blocking)
# Добавление ключей блокировки по дате рождения
dataset1['blocking_key'] = dataset1['birthdate']
dataset2['blocking_key'] = dataset2['birthdate']
dataset3['blocking_key'] = dataset3['birthdate']
```

```python
from joblib import Parallel, delayed

# Параллелизация
def compute_features(row1, row2, row3):
    return create_features(row1, row2, row3)

# Параметр n_jobs=-1 использует все доступные ядра процессора
features_list = Parallel(n_jobs=-1)(
    delayed(compute_features)(row1, row2, row3)
    for i, row1 in dataset1.iterrows()
    for j, row2 in dataset2.iterrows()
    for k, row3 in dataset3.iterrows()
)

features_df = pd.DataFrame(features_list)
features_df
```

```python
from rapidfuzz import fuzz

def create_features(row1, row2, row3):
    features = {}
    features['name_similarity'] = fuzz.token_sort_ratio(
        row1['full_name'],
        " ".join([row2['first_name'], row2['middle_name'], row2['last_name']])
    )
    features['email_similarity'] = fuzz.ratio(row1['email'], row3['email'])
    features['name_similarity_3'] = fuzz.ratio(row1['full_name'], row3['name'])
    features['email_similarity_3'] = fuzz.ratio(row1['email'], row3['email'])
    return features

# Пример создания признаков для пар записей
features_list = []
for i, row1 in dataset1.iterrows():
    for j, row2 in dataset2.iterrows():
        for k, row3 in dataset3.iterrows():
            features = create_features(row1, row2, row3)
            features['index1'] = i
            features['index2'] = j
            features['index3'] = k
            features_list.append(features)

features_df = pd.DataFrame(features_list)
features_df
```

```python
# Шаг 4: Сопоставление записей
# Устанавливаем порог для сходства
similarity_threshold = 85

# Находим совпадения
matches = features_df[
    (features_df['name_similarity'] >= similarity_threshold) |
    (features_df['email_similarity'] >= similarity_threshold) |
    (features_df['name_similarity_3'] >= similarity_threshold) |
    (features_df['email_similarity_3'] >= similarity_threshold)
]
```

## Оценка качества

```python
# Шаг 6: Объединение совпавших записей
def merge_records(row1, row2, row3):
    merged = {
        'full_name': row1['full_name'],
        'email': row1['email'] if row1['email'] else (row2['email'] if row2['email'] else row3['email']),
        'address': row1['address'] if row1['address'] else row2['address'],
        'phone': row1['phone'] if row1['phone'] else row2['phone'],
        'birthdate': row1['birthdate'] if row1['birthdate'] else (row2['birthdate'] if row2['birthdate'] else row3['birthdate'])
    }
    return merged

# Объединяем совпадения
final_records = []
for _, match in matches.iterrows():
    row1 = dataset1.iloc[int(match['index1'])]
    row2 = dataset2.iloc[int(match['index2'])]
    row3 = dataset3.iloc[int(match['index3'])]
    final_records.append(merge_records(row1, row2, row3))

# Создаем итоговый DataFrame
final_df = pd.DataFrame(final_records)
final_df
```

## Машинное обучение

```python
import pandas as pd
import re
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.metrics import precision_score, recall_score, f1_score

# Шаг 4: Сопоставление записей с помощью DBSCAN
X = features_df[['name_similarity', 'email_similarity', 'name_similarity_3', 'email_similarity_3']]

# Нормализация данных
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Настраиваем и применяем DBSCAN
dbscan = DBSCAN(eps=1.0, min_samples=2)
features_df['cluster'] = dbscan.fit_predict(X_scaled)

# Фильтруем шум (-1) и стягиваем записи по кластерам
final_records = []
clusters = features_df[features_df['cluster'] != -1]

for cluster_id in clusters['cluster'].unique():
    cluster_data = clusters[clusters['cluster'] == cluster_id]

    # Берем индексы записей из всех трех датасетов
    indices1 = cluster_data['index1'].unique()
    indices2 = cluster_data['index2'].unique()
    indices3 = cluster_data['index3'].unique()

    row1 = dataset1.iloc[indices1[0]] if len(indices1) > 0 else None
    row2 = dataset2.iloc[indices2[0]] if len(indices2) > 0 else None
    row3 = dataset3.iloc[indices3[0]] if len(indices3) > 0 else None

    final_record = merge_records(row1, row2, row3)
    final_records.append(final_record)

# Создаем итоговый DataFrame
final_df = pd.DataFrame(final_records)
final_df
```

**OPTIMIZATION**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.metrics import precision_score, recall_score, f1_score
from IPython.display import display

# Генерация случайных истинных меток
true_labels = np.random.randint(0, 2, size=len(features_df))  # Генерация случайных меток
features_df['true_labels'] = true_labels

# Шаг 4: Сопоставление записей с помощью DBSCAN
X = features_df[['name_similarity', 'email_similarity', 'name_similarity_3', 'email_similarity_3']]

# Нормализация данных
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Настраиваем и применяем DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=2)
features_df['cluster'] = dbscan.fit_predict(X_scaled)

# Фильтруем шум (-1) и стягиваем записи по кластерам
final_records = []
clusters = features_df[features_df['cluster'] != -1]

for cluster_id in clusters['cluster'].unique():
    cluster_data = clusters[clusters['cluster'] == cluster_id]

    indices1 = cluster_data['index1'].unique()
    indices2 = cluster_data['index2'].unique()
    indices3 = cluster_data['index3'].unique()

    row1 = dataset1.iloc[indices1[0]] if len(indices1) > 0 else None
    row2 = dataset2.iloc[indices2[0]] if len(indices2) > 0 else None
    row3 = dataset3.iloc[indices3[0]] if len(indices3) > 0 else None

    final_record = merge_records(row1, row2, row3)
    final_records.append(final_record)


# Создаем итоговый DataFrame и удаляем дубликаты
final_df = pd.DataFrame(final_records).drop_duplicates().reset_index(drop=True)

# Вычисление метрик
y_true = features_df['true_labels']
y_pred = features_df['cluster']

# Вычисляем метрики
precision = precision_score(y_true, y_pred, average='weighted', zero_division=0)
recall = recall_score(y_true, y_pred, average='weighted', zero_division=0)
f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)

print(f'Precision: {precision}, Recall: {recall}, F1 Score: {f1}')


# Выводим итоговый DataFrame
display(final_df)
```
